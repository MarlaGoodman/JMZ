```python
import pandas as pd
```

```python
df=pd.read_csv('/Users/marleegoodman/Dropbox/My Mac (Marleeâ€™s MacBook Air)/Desktop/RG.csv')
pd.set_option('display.max_columns', None)
pd.set_option("max_rows", None)
dft=df.T
```

```python
new_header = dft.iloc[0] #grab the first row for the header
dft = dft[1:] #take the data less the header row
dft.columns = new_header #set the header row as the df header
dft
```

```python
import matplotlib.pyplot as plt
```

```python
dfrg=dft.apply(pd.Series.value_counts)
dfrg
```

```python
plt.figure(figsize=(10, 10))
dfrg.iloc[0].sort_values(ascending=False).plot(kind='bar',title='Green Amounts')
plt.ylabel('Numb. of Greens')
plt.xlabel('Employee')
plt.show()
```

```python
perc=dfrg.iloc[0]/29
plt.figure(figsize=(10, 10))
perc.sort_values(ascending=False).plot(kind='bar',title='Green Percentage')
plt.ylabel('Percent of Greens')
plt.xlabel('Employee')
plt.show()
```
